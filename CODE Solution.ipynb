```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("C:/Users/User/User/Desktop/CSV.csv")
data['month']=pd.DatetimeIndex(data['date_added']).day
data
```

```python
ans = pd.DataFrame()
```

```python
Data = np.zeros((data['customer_id'].max(),12,2))
for i in set(data['customer_id']):
    month = np.zeros((12,2))
    df = data[data['customer_id'] == i]
    for j in (df.index.get_values()):
        if df.collection[j] == 'payable':
            month[df.month[j]-1][0] = df.amount[j]
        if df.collection[j] == 'paid':
            month[df.month[j]-1][1] = df.amount[j]
    Data[i-1] = month
print(Data)
```

```python
ans['customer_id'] = data['customer_id'].unique()
for i in range(11):
    payable1 = pd.Series((Data[:,i+1,0]).tolist())
    paid1 = pd.Series((Data[:,i+1,1]).tolist())
    ans['month '+ str(i+2)+ ' Payable'] = payable1.values
    ans['month '+str(i+2) +' Paid'] = paid1.values
```

```python
NetPay = []
NetPaid = []
for i in range(data['customer_id'].unique().max()):
    NetPay.append(Data[i,:,0].sum())
    NetPaid.append(Data[i,:,1].sum())

NetTotal = np.array(NetPaid) - np.array(NetPay)
Net_Total = NetTotal.tolist()
ans['Net_Payable'] = pd.Series(NetPay).values
ans['Net_Paid'] = pd.Series(NetPaid).values
ans['Net_total'] = pd.Series(Net_Total).values
ans.to_csv('C:/Users/User/User/Desktop/Answer',sep='\t')
```
